```python
class ConnectFourBoard:
    def __init__(self, setup_of_board = "New"):
        self.X = "X"
        self.O = "O"
        self.space = " "
        self.draw = 'D'
        self.board_width = 7
        self.board_height = 6
        self.board_size = self.board_width * self.board_height
    
        if setup_of_board == "New":
            self.board = self.space * self.board_size
        elif len(setup_of_board) != self.board_size:
            print("Invalid Setup Error, Length expected: {}, Length provided: {}".format(self.board_size, len(setup_of_board)))
        else:
            self.board = setup_of_board

    
    def display(self, board_string = ""):
        for row_index in range(self.board_height):
            current_row = ''
            for column_index in range(self.board_width):
                pos = (row_index * self.board_width) + column_index
                current_row = current_row + self.board[pos]
            print(current_row)
            
        print("-" * self.board_width)
        print(string.digits[:self.board_width])
        
    def on_same_row(self, first_position, second_position):
        if(first_position == second_position):
            return False
        elif(first_position > second_position):
            return self.on_same_row(second_position, first_position)
        elif(second_position >= self.board_size or first_position >= self.board_size):
            print("Invalid Check, first or second position greater than board size")
            return False
        elif(first_position% self.board_width >= second_position% self.board_width or (second_position-first_position) >= self.board_width):
            return False
        else: 
            return True
    
    def on_same_column(self, first_position, second_position):
        return(first_position + self.board_width == second_position or first_position - self.board_width == second_position)
    
    def is_on_edge_left(self, position):
        return position % self.board_width == 0
    
    def is_on_edge_right(self, position):
        return position % self.board_width == self.board_width - 1
        
    def game_won_diagonal_left_right(self, position):
        count = 1
        current_position_a = position
        current_position_b = position
        
        while count < 4:
            left_edge = self.is_on_edge_left(current_position_a)
            right_edge = self.is_on_edge_right(current_position_b)
            
            if not left_edge:
                new_position = current_position_a - (self.board_width + 1)
                if self.equal_at(current_position_a, new_position):
                    count = count + 1
                    current_position_a = new_position
            
            if not right_edge:
                new_position = current_position_b + (self.board_width + 1)
                if self.equal_at(current_position_b, new_position):
                    count = count + 1
                    current_position_b = new_position
                
            if left_edge and right_edge:
                break
                
        if count >= 4:
            return True
        else:
            return False
    
    def game_won_diagonal_right_left(self, position):
        count = 1
        current_position_a = position
        current_position_b = position
        
        while count < 4:
            left_edge = self.is_on_edge_left(current_position_a)
            right_edge = self.is_on_edge_right(current_position_b)
            
            if not left_edge:
                new_position = current_position_a + (self.board_width - 1)
                if self.equal_at(current_position_a, new_position):
                    count = count + 1
                    current_position_a = new_position
            
            if not right_edge:
                new_position = current_position_b - (self.board_width - 1)
                if self.equal_at(current_position_b, new_position):
                    count = count + 1
                    current_position_b = new_position
                
            if left_edge and right_edge:
                break
                
        if count >= 4:
            return True
        else:
            return False
        
    def game_won_diagonal(self, position):
        return self.game_won_diagonal_left_right(position) or self.game_won_diagonal_right_left(position)
    
    def no_more_spaces(self):
        for i in range(self.board_size):
            if(self.board[i] == self.space):
                return self.space
        return self.draw
    
    def game_won_horizontal(self, position):
        count = 1
        current_position_r = position
        current_position_l = position
        
        valid_right = self.is_valid_position(current_position_r)
        valid_left = self.is_valid_position(current_position_l)
            
        while count < 4:
            # check pieces to the right
            if valid_right:
                new_position = current_position_r+1
                if self.on_same_row(current_position_r, new_position) and self.equal_at(current_position_r, new_position):
                    count = count + 1
                    current_position_r = new_position
                else:
                    valid_right = False
            
            # check pieces to the left
            if valid_left:
                new_position = current_position_l-1
                if self.on_same_row(current_position_l, new_position) and self.equal_at(current_position_l, new_position):
                    count = count + 1
                    current_position_l = new_position
                else:
                    valid_left = False
        
        if count >= 4:
            return True
        else:
            return False
        
    def game_won_vertical(self, position):
        count = 1
        current_position_u = position
        current_position_d = position
        
        valid_up = self.is_valid_position(current_position_u)
        valid_down = self.is_valid_position(current_position_d)
            
        while count < 4:
            # check pieces above
            if valid_up:
                new_position = current_position_u+self.board_width
                if self.equal_at(current_position_u, new_position):
                    count = count + 1
                    current_position_u = new_position
                else:
                    valid_up = False
            
            # check pieces below
            if valid_down:
                new_position = current_position_d-self.board_width
                if self.equal_at(current_position_d, new_position):
                    count = count + 1
                    current_position_d = new_position
                else:
                    valid_down = False
        
        if count >= 4:
            return True
        else:
            return False
        
    def is_game_over(self, position):
        if(self.board[position] == self.space):
            return False
        elif((self.game_won_horizontal(position)) or (self.game_won_vertical(position)) or (self.game_won_diagonal(position))):
            return True
        else:
            return False
          
    def game_won(self):
        print('test')
        for i in range(self.board_size):
            print(i)
            if self.board[i] != self.space and self.is_game_over(i):
                return self.board[i]
        return self.no_more_spaces()
        
    def other(self, character):
        if character == self.X:
            return self.O
        elif character == self.O:
            return self.X
        else:
            print('invalid character chosen {}'.format(character))
    
    def is_valid_position(self, position):
        return position >= 0 and position < self.board_size
    
    def equal_at(self, start_position, new_position):
        return self.is_valid_position(start_position) and self.is_valid_position(new_position) and self.board[start_position] == self.board[new_position]
    
    def drop_to_index(self, current_index):
        next_index_position = current_index + self.board_width
        if (next_index_position < self.board_size) and (self.board[next_index_position] == self.space):
            return self.drop_to_index(next_index_position)
        else:
            return current_index
    
    
    #Result * 1/ (Plydepth + 1)
    def all_possible_moves_for(self, choosing_this_character):
        #returns (as GameStates) a list of all possible moves
        list_of_all_moves = []
        for column in range(self.board_width):
            index = self.drop_to_index(column)
            list_of_all_moves.append(ConnectFourBoard(self.board[:index]+choosing_this_character+self.board[index+1:]))
        return list_of_all_moves
```

```python
class SearchTreeNode():
    def __init__(self, just_after_choosing_character, current_game = None, Ply_Depth = 0):
        self.children = []
        self.ply_depth = Ply_Depth
        self.is_node_value_assigned = False
        self.node_value = 0
        self.choosing_character = just_after_choosing_character
        self.max_ply_depth = 7
        
        if current_game == None:
            self.game_at_this_node = ConnectFourBoard()
        else:
            self.game_at_this_node = current_game
            
        self.generate_children()
        
    def generate_children(self):
        if self.game_at_this_node.game_won() != self.game_at_this_node.space:
            #This is a leaf, apply utility value function
            if (self.ply_depth) % 2 == 1:
                self.node_value = 1
            else:
                self.node_value = -1
            self.is_node_value_assigned = True
        elif self.ply_depth >= self.max_ply_depth:
            self.is_node_value_assigned = True
        else:
            all_moves = self.game_at_this_node.all_possible_moves_for(self.game_at_this_node.other(self.choosing_character))
            if (len(all_moves) == 0):
                self.node_value = 0
                self.is_node_value_assigned = True
            else:
                for new_game in all_moves:
                    print('a')
                    self.children.append(SearchTreeNode(new_game.other(self.choosing_character), new_game, self.ply_depth+1))
        
    def get_node_value(self):
        if not self.is_node_value_assigned:
            self.is_node_value_assigned = True
            self.children = sorted(self.children, key = lambda x: x.get_node_value())
            
            if (self.ply_depth % 2) == 0: # Maximising Layer
                self.node_value = self.children[-1].node_value
            else: # Minimising Layer
                self.node_value = self.children[0].node_value
            
        return self.node_value
```

```python
def play_connect_four():
    current_board = None
    player_choosing = ""
    computer_choosing = ""
    choice = input("Do you want to go first? y/n ")
    player_going_first = (choice == "y" or choice == "Y")
    choice = input("Do you want to play as X? y/n ")
    if choice == "y" or choice == "Y":
        player_choosing = "X"
        computer_choosing = "O"
    else:
        player_choosing = "O"
        computer_choosing = "X"
    if player_going_first:    
        current_board = ConnectFourBoard()
    else:
        #Choose a random move
        current_board = ConnectFourBoard()
        all_current_moves = current_board.all_possible_moves_for(computer_choosing)
        current_board = all_current_moves[np.random.randint(len(all_current_moves))]
        #current_board = ConnectFourBoard(current_board.board[:ind] + computer_choosing + current_board.board[ind+1:])
        
    for i in range(42):
        current_board.display()
        index_as_string = input("Choose a row to place your next piece ")
        ind = int(index_as_string)
        #prevents invalid input
        ind = ind % 7
        ind = current_board.drop_to_index(ind)
        current_board = ConnectFourBoard(current_board.board[:ind] + player_choosing + current_board.board[ind+1:])
        current_board.display()
        if current_board.game_won() == player_choosing:
            print("I can't believe you've won!!")
            current_board.display()
            break
        elif current_board.no_more_spaces() == current_board.draw:
            print("this is a draw")
            current_board.display()
            break
        search_tree = SearchTreeNode(player_choosing, current_board)
        search_tree.get_node_value()
        current_board = search_tree.children[-1].game_at_this_node
        if current_board.game_won() == computer_choosing:
            print("Haha you lost loser")
            current_board.display()
            break
        elif current_board.no_more_spaces() == current_board.draw:
            print("issa draw")
            current_board.display()
            break
        
    return player_going_first, player_choosing, current_board
```
